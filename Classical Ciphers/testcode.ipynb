```python
import random
import textwrap
from collections import OrderedDict
```

```python
class Cipher:
    check = False
    def __init__(self, x):
        self.decrypted_message=x
        self.encrypted_message=""
    
    def get_Check_Decryption(self):
        answer_for_decryption_check = input("Do you want to check the decryption? (y/n): ")
        if answer_for_decryption_check == "y":
            self.check = True
        else:
            self.check = False
        
```

```python
class Caeser_cipher_v1(Cipher):
    key = 0
    
    def get_Key(self):
        #Ask the user for a key number
        self.key = int(input("Type the key for this message (a number): ")) % 26
    
    def Encrypt(self):
        
        #Initiate the Key from the user
        self.get_Key()
        
        #Initiate the encrypted message variable
        self.encrypted_message = ""
        
        #Split the text in sentences
        sentences = self.decrypted_message.split(".")
        
        #Delete the last empty string
        if not sentences[len(sentences)-1]:
            sentences = sentences[:-1]
        words_in_sentences = []
        
        #Split each sentence into a list of words
        for sentence in sentences:
            words_in_sentences.append(sentence.split())
            
        #Iterate through each sentence, each word and each letter
        for sentence in words_in_sentences:
            for word in sentence:
                for letter in word:
                    #Check if upper or lower case
                    if letter.isupper():
                        #In ASCII the Upper letters start from 65
                        limit = 65
                    else:
                        #In ASCII the Lower letters start from 97
                        limit = 97
                    
                    #Step 1: Get the number of the character (1-A ... 26-Z)
                    char_number = ord(letter) - limit
                    #Step 2: Update it with key numbers forward
                    char_number = (char_number + self.key) % 26
                    #Step 3: Transform it into a letter
                    char = chr(char_number + limit)
                    #Step 4: Add it to the Crypted Message
                    self.encrypted_message += char
                    
                #Adding space after encrypted words
                self.encrypted_message += " "
            
            #Deleting the last space
            self.encrypted_message = self.encrypted_message.rstrip()
            #Adding ". " after the encrypted sentence
            self.encrypted_message += ". "
            
            print("The original message:\n"+self.decrypted_message+"\nThe encrypted message:\n"+self.encrypted_message+"\n")
            
    def Decrypt(self):
        
        #Initiate the decrypted message value to empty string
        self.decrypted_message = ""
        
        #Split the text in sentences
        sentences = self.encrypted_message.split(".")
        
        #Delete the last empty string
        sentences = sentences[:-1]
        
        words_in_sentences = []
        
        #Split each sentence into a list of words
        for sentence in sentences:
            words_in_sentences.append(sentence.split())
            
        #Iterate through each sentence, each word and each letter
        for sentence in words_in_sentences:
            for word in sentence:
                for letter in word:
                    #Check if upper or lower case
                    if letter.isupper():
                        #In ASCII the Upper letters start from 65
                        limit = 65
                    else:
                        #In ASCII the Lower letters start from 97
                        limit = 97
                        
                    #Step 1: Get the number of the character (1-A ... 26-Z)
                    char_number = ord(letter) - limit
                    #Step 2: Update it with key numbers backwards
                    char_number = (char_number + (26 - self.key)) % 26
                    #Step 3: Transform it into a letter
                    char = chr(char_number + limit)
                    #Step 4: Add it to the Decrypted Message
                    self.decrypted_message += char
                    
                #Adding space after decrypted words
                self.decrypted_message += " "
            
            #Deleting the last space
            self.decrypted_message = self.decrypted_message.rstrip()
            #Adding ". " after the decrypted sentence
            self.decrypted_message += ". "
            
        print(f"The encrypted message:\n"+self.encrypted_message+"\nThe decrypted message:\n"+self.decrypted_message)
```

```python
class Vigenere_cipher(Cipher):
    key = []
    
    def get_Key(self):
        
        #Ask the user for the Vigenere key
        word_key = input("Type the key for this message (a word): ")
        
        #Define the number of letters in the key (starts form 0)
        self.key_length = len(word_key)-1
        
        #transform all the letters to lowercases
        word_key = word_key.lower()
        
        #Initiate the key list
        self.key = []
        
        #Add the number value of the letters to the key area
        for letter in word_key:
            self.key.append(ord(letter)-97+1)
                
    
    def Encrypt(self):
        
        
        #Initiate the Key from the user
        self.get_Key()
        
        #Initiate the encrypted message variable
        self.encrypted_message = ""
        
        #Split the text in sentences
        sentences = self.decrypted_message.split(".")
        
        #Delete the last empty string
        if not sentences[len(sentences)-1]:
            sentences = sentences[:-1]
        words_in_sentences = []
        
        #Split each sentence into a list of words
        for sentence in sentences:
            words_in_sentences.append(sentence.split())
            
        #Define the key iterative
        key_i = 0
        #Iterate through each sentence, each word and each letter
        for sentence in words_in_sentences:
            for word in sentence:
                for letter in word:
                    #Check if upper or lower case
                    if letter.isupper():
                        #In ASCII the Upper letters start from 65
                        limit = 65
                    else:
                        #In ASCII the Lower letters start from 97
                        limit = 97
                    
                    #Step 1: Get the number of the character (1-A ... 26-Z)
                    char_number = ord(letter) - limit
                    #Step 2.1: Update it with key[key_i] numbers forward
                    char_number = (char_number + self.key[key_i]) % 26
                    #Step 2.2: Update the key_i
                    key_i = (key_i + 1) % self.key_length 
                    #Step 3: Transform it into a letter
                    char = chr(char_number + limit)
                    #Step 4: Add it to the Crypted Message
                    self.encrypted_message += char
                    
                #Adding space after encrypted words
                self.encrypted_message += " "
            
            #Deleting the last space
            self.encrypted_message = self.encrypted_message.rstrip()
            #Adding ". " after the encrypted sentence
            self.encrypted_message += ". "
            print("The original message:\n"+self.decrypted_message+"\nThe encrypted message:\n"+self.encrypted_message+"\n")
            
    def Decrypt(self):
        
        #Initiate the decrypted message value to empty string
        self.decrypted_message = ""
        
        #Split the text in sentences
        sentences = self.encrypted_message.split(".")
        
        #Delete the last empty string
        if not sentences[len(sentences)-1]:
            sentences = sentences[:-1]
        
        words_in_sentences = []
        
        #Split each sentence into a list of words
        for sentence in sentences:
            words_in_sentences.append(sentence.split())
            
        #Define the key iterative
        key_i = 0
        #Iterate through each sentence, each word and each letter
        for sentence in words_in_sentences:
            for word in sentence:
                for letter in word:
                    #Check if upper or lower case
                    if letter.isupper():
                        #In ASCII the Upper letters start from 65
                        limit = 65
                    else:
                        #In ASCII the Lower letters start from 97
                        limit = 97
                        
                    #Step 1: Get the number of the character (1-A ... 26-Z)
                    char_number = ord(letter) - limit
                    #Step 2.1: Update it with key[key_i] numbers backwards
                    char_number = (char_number + (26 - self.key[key_i])) % 26
                    #Step 2.2: Update the key_i
                    key_i = (key_i + 1) % self.key_length
                    #Step 3: Transform it into a letter
                    char = chr(char_number + limit)
                    #Step 4: Add it to the Decrypted Message
                    self.decrypted_message += char
                    
                #Adding space after decrypted words
                self.decrypted_message += " "
            
            #Deleting the last space
            self.decrypted_message = self.decrypted_message.rstrip()
            #Adding ". " after the decrypted sentence
            self.decrypted_message += ". "
            
        print(f"The encrypted message:\n"+self.encrypted_message+"\nThe decrypted message:\n"+self.decrypted_message)
```

```python
class Caeser_cipher_v2(Cipher):
    key = 0
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    
    def get_Key(self):
        #Ask the user for a key number
        self.key = int(input("Type the key for this message (a number): ")) % 26
        self.get_Alphabet()
        
    def get_Alphabet(self):
        self.alphabet = 'abcdefghijklmnopqrstuvwxyz'
        self.alphabet = ''.join(random.sample(self.alphabet,26))
        print("\nThe secret random alphabet:\n"+self.alphabet+"\n")
    
    def Encrypt(self):
        
        #Initiate the Key from the user
        self.get_Key()
        
        #Initiate the encrypted message variable
        self.encrypted_message = ""
        
        #Split the text in sentences
        sentences = self.decrypted_message.split(".")
        
        #Delete the last empty string
        if not sentences[len(sentences)-1]:
            sentences = sentences[:-1]
        words_in_sentences = []
        
        #Split each sentence into a list of words
        for sentence in sentences:
            words_in_sentences.append(sentence.split())
            
        #Iterate through each sentence, each word and each letter
        for sentence in words_in_sentences:
            for word in sentence:
                for letter in word:
                    #Set the letter being upper to false
                    Upper = False
                    #Check if upper or lower case
                    if letter.isupper():
                        #In ASCII the Upper letters start from 65
                        letter = letter.lower()
                        Upper = True
                    
                    #Step 1: Get the number of the character from the random alphabet
                    char_number = self.alphabet.find(letter)
                    #Step 2: Update it with key numbers forward
                    char_number = (char_number + self.key) % 26
                    #Step 3.1: Transform it into a letter from the random alphabet
                    char = self.alphabet[char_number]
                    #Step 3.2: Transform it into an upper if needed
                    if Upper:
                        char = char.upper()
                    #Step 4: Add it to the Crypted Message
                    self.encrypted_message += char
                    
                #Adding space after encrypted words
                self.encrypted_message += " "
            
            #Deleting the last space
            self.encrypted_message = self.encrypted_message.rstrip()
            #Adding ". " after the encrypted sentence
            self.encrypted_message += ". "
            
            print("The original message:\n"+self.decrypted_message+"\nThe encrypted message:\n"+self.encrypted_message+"\n")
            
    def Decrypt(self):
        
        #Initiate the decrypted message value to empty string
        self.decrypted_message = ""
        
        #Split the text in sentences
        sentences = self.encrypted_message.split(".")
        
        #Delete the last empty string
        if not sentences[len(sentences)-1]:
            sentences = sentences[:-1]
        
        words_in_sentences = []
        
        #Split each sentence into a list of words
        for sentence in sentences:
            words_in_sentences.append(sentence.split())
            
        #Iterate through each sentence, each word and each letter
        for sentence in words_in_sentences:
            for word in sentence:
                for letter in word:
                    #Set the letter being upper to false
                    Upper = False
                    #Check if upper or lower case
                    if letter.isupper():
                        #In ASCII the Upper letters start from 65
                        letter = letter.lower()
                        Upper = True
                        
                    #Step 1: Get the number of the character from the random alphabet
                    char_number = self.alphabet.find(letter)
                    #Step 2: Update it with key numbers backwards
                    char_number = (char_number + (26 - self.key)) % 26
                    #Step 3.1: Transform it into a letter from the random alphabet
                    char = self.alphabet[char_number]
                    #Step 3.2: Transform it into an upper if needed
                    if Upper:
                        char = char.upper()
                    #Step 4: Add it to the Decrypted Message
                    self.decrypted_message += char
                    
                #Adding space after decrypted words
                self.decrypted_message += " "
            
            #Deleting the last space
            self.decrypted_message = self.decrypted_message.rstrip()
            #Adding ". " after the decrypted sentence
            self.decrypted_message += ". "
            
        print(f"The encrypted message:\n"+self.encrypted_message+"\nThe decrypted message:\n"+self.decrypted_message)
```

```python
class Playfair_cipher(Cipher):
    key = ''
    alphabet = "ABCDEFGHIKLMNOPQRSTUVWXYZ"
    #A dictionary to save the letter for each code position
    Pos_to_letter = {}
    #A dictionary to save the code position for each letter
    Letter_to_pos = {}
    #The bogus letter we will add to the mono-letter duos
    bogus_letter = "S"
    
    #Transfor the Key into a sequence of unique letters, ordered in their appearance
    def Transform_to_Unique(self, S):
        return "".join(OrderedDict.fromkeys(S))
    
    #Add the Key and the alphabet to the Dictionaries with the respective codes
    def Key_to_Dictionaries(self):
        #Dictionary with a letter for each position
        self.Pos_to_letter = {}
        #Dictionary with a position for each letter
        self.Letter_to_pos = {}
        
        #Update the key with the missing letters from the alphabet
        for letter in self.alphabet:
            if letter not in self.key:
                self.key += letter
                
        # i -> counter for row
        i = 0
        # j -> counter for column
        j = 0
        
        for letter in self.key:
            #define the code for the letter
            cod_string = str(i)+str(j)
            #Update the counters
            j+=1
            if j==5:
                j=0
                i+=1  
            
            #Atribute to the code the respective letter
            self.Pos_to_letter[cod_string] = letter
            #Atribute to the letter the respective code
            self.Letter_to_pos[letter] = cod_string
        
        #Print the alphabet and their code
        for key in self.Letter_to_pos:
            print(key+" -> "+self.Letter_to_pos[key])
    
    def get_Key(self):
        #Ask the user for a key number
        key_word = input("Type the key for this message (a word): ")
        #Transform the Key to upper letters
        key_word = key_word.upper()
        #Transform the Key into a sequence of unique letters
        self.key = self.Transform_to_Unique(key_word)
        #Populate the dictionaries with letters and their respective codes
        self.Key_to_Dictionaries()
        
            
    def Encrypt(self):
        
        #Initiate the Key from the user
        self.get_Key()
        
        #Initiate the encrypted message variable
        self.encrypted_message = ""
        
        #Split the text in sentences
        sentences = self.decrypted_message.split(".")
        
        #Delete the last empty string
        if not sentences[len(sentences)-1]:
            sentences = sentences[:-1]
        words_in_sentences = []
        
        #Split each sentence into a list of words
        for sentence in sentences:
            words_in_sentences.append(sentence.split())
        
        #Duos are groups of 2 letters
        duos = []
        new_duo = ""
        #Preparing the list of duos
        for sentence in words_in_sentences:
            for word in sentence:
                new_duo = ""
                #Step 1: Change the letters to upper letters
                word = word.upper()
                #Step 2: Replace the J with I
                word = word.replace("J","I")
                #Step 3: divide into duos correctly according to scenarios
                for letter in word:
                    #if the new duo has no letters
                    if not new_duo:
                        #Just add the current letter
                        new_duo = new_duo + letter
                        
                    # else if it does
                    elif new_duo:
                        #if the letter is already present in the new duo
                        if letter in new_duo:
                            #Add the single duo to the list
                            duos.append(new_duo)
                            #Update the new duo to being the current letter
                            new_duo = letter
                        #else if the letter is not present in the new duo
                        else:
                            #Add the letter to the new duo
                            new_duo = new_duo + letter
                            #Add the duo to the list
                            duos.append(new_duo)
                            #Reinitialize the variable
                            new_duo = ""
                #if the variable still hold a letter, add it to the list, it is the last letter of the text            
                if new_duo:
                    duos.append(new_duo)
                    
        #Initialise the list of the encrypted duos
        encrypted_duos = []
        
        #Encryption starts here
        for i in range(len(duos)):
            
            #if the duo has only one letter add a bogus letter
            if len(duos[i]) == 1:
                duos[i] = duos[i] + self.bogus_letter
            
            #Define the letters and their codes
            letter_1 = duos[i][0]
            letter_2 = duos[i][1]
            code_1 = self.Letter_to_pos[letter_1]
            code_2 = self.Letter_to_pos[letter_2]
            
            #Case 1: If the letters are on the same row 
            if code_1[0] == code_2[0]:
                #Increase the column number with one
                change_int_2_1 = (int(code_1[1])+1)%5
                change_int_2_2 = (int(code_2[1])+1)%5
                code_1 = code_1[0] + str(change_int_2_1)
                code_2 = code_2[0] + str(change_int_2_2)
                encrypted_duo = self.Pos_to_letter[code_1] + self.Pos_to_letter[code_2]
            
            #Case 2: If the letters are on the same column
            elif code_1[1] == code_2[1]:
                #Increase the row number with one
                change_int_1_1 = (int(code_1[0])+1)%5
                change_int_1_2 = (int(code_2[0])+1)%5
                code_1 = str(change_int_1_1) + code_1[1]
                code_2 = str(change_int_1_2) + code_2[1]
                encrypted_duo = self.Pos_to_letter[code_1] + self.Pos_to_letter[code_2]
            
            #Case 3: If they are in the oposite corners of a square
            else:
                #Change the number of columns between them = moving to the other corner
                change_int_2_1 = int(code_2[1])
                change_int_2_2 = int(code_1[1])
                code_1 = code_1[0] + str(change_int_2_1)
                code_2 = code_2[0] + str(change_int_2_2)
                encrypted_duo = self.Pos_to_letter[code_1] + self.Pos_to_letter[code_2]
                
            encrypted_duos.append(encrypted_duo)
        
        #Glue the duos in one encrypted message
        for duo in encrypted_duos:
            self.encrypted_message += duo
        
        print("The original message:\n"+self.decrypted_message+"\nThe encrypted message:\n"+self.encrypted_message+"\n")
            
    def Decrypt(self):
        
        #Initiate the decrypted message value to empty string
        self.decrypted_message = ""
        
        duos = []
        duos += textwrap.wrap(self.encrypted_message,2)
        
        #Initialise the list of the encrypted duos
        decrypted_duos = []
        
        #Decryption starts here
        for i in range(len(duos)):
            
            decrypted_duo = ""
            
            #Define the letters and the duos codes
            letter_1 = duos[i][0]
            letter_2 = duos[i][1]
            code_1 = self.Letter_to_pos[letter_1]
            code_2 = self.Letter_to_pos[letter_2]
            
            #Case 1: If the letters are on the same row 
            if code_1[0] == code_2[0]:
                #Decrease the column number with one
                change_int_2_1 = abs((int(code_1[1])-1))
                change_int_2_2 = abs((int(code_2[1])-1))
                code_1 = code_1[0] + str(change_int_2_1)
                code_2 = code_2[0] + str(change_int_2_2)
                decrypted_duo = self.Pos_to_letter[code_1] + self.Pos_to_letter[code_2]
                
            #Case 2: If the letters are on the same column
            elif code_1[1] == code_2[1]:
                #Decrease the row number with one
                change_int_1_1 = abs((int(code_1[0])-1))
                change_int_1_2 = abs((int(code_2[0])-1))
                code_1 = str(change_int_1_1) + code_1[1]
                code_2 = str(change_int_1_2) + code_2[1]
                decrypted_duo = self.Pos_to_letter[code_1] + self.Pos_to_letter[code_2]
                
            #Case 3: If they are in the oposite corners of a square
            else:
                #Change the number of columns between them = moving to the other corner
                change_int_2_1 = int(code_2[1])
                change_int_2_2 = int(code_1[1])
                code_1 = code_1[0] + str(change_int_2_1)
                code_2 = code_2[0] + str(change_int_2_2)
                decrypted_duo = self.Pos_to_letter[code_1] + self.Pos_to_letter[code_2]
                
            decrypted_duos.append(decrypted_duo)
            
        #Glue the duos in one decrypted message
        for duo in decrypted_duos:
            self.decrypted_message += duo
            
        #Delete the bogus letter
        self.decrypted_message =  self.decrypted_message.replace(self.bogus_letter,"")
        
        print("The encrypted message:\n"+self.encrypted_message+"\nThe decrypted message:\n"+self.decrypted_message)
```

```python
def starting():
    
    plain_text = str(input("Please introduce the text you would like to encrypt: "))
    print("\nConsider the following classical ciphers:")
    print("1 - Caesar classic cipher")
    print("2 - Caesar cipher with random alphabet")
    print("3 - Vigenere cipher")
    print("4 - Playfair cipher")

    type_cipher = int(input("\nPlease introduce the index of the cipher you would like to use: "))

    if type_cipher == 1:
        Encryption = Caeser_cipher_v1(plain_text)
    elif type_cipher == 2:
        Encryption = Caeser_cipher_v2(plain_text)
    elif type_cipher == 3:
        Encryption = Vigenere_cipher(plain_text)
    else:
        Encryption = Playfair_cipher(plain_text)

    Encryption.Encrypt()
    
    Encryption.get_Check_Decryption()

    if Encryption.check:
        Encryption.Decrypt()
    return
```
